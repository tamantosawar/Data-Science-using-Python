he fillna() method, which returns a copy of the array with the null values
replaced.

```python
data = pd.Series([1, np.nan, 2, None, 3], index=list('abcde'))
data
```

```python
data.fillna(0)
```

```python
data.fillna(method='ffill') #ffill bfill
```

For DataFrames, the options are similar, but we can also specify an axis along which
the fills take place:

```python
df = pd.DataFrame([[1, np.nan, 2],
 [2, 3, 5],
 [np.nan, 4, 6]])
```

```python
df
```

```python
df.fillna(method='ffill', axis=1)
```
